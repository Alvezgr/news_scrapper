# Data Wrangling

```python
# 1. add newspeper_uid to the dataframe
el_universal['newspaper_uid'] = 'eluniversal'
el_universal
```

```python
# 2. Getting the host
from urllib.parse import urlparse

el_universal['host'] = el_universal['uri'].apply(lambda uri: urlparse(uri).netloc)
el_universal
```
