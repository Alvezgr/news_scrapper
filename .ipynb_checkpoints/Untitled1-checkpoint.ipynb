```python
import requests
import bs4
titulos = []
for i in range(0,31):
    if i<10:
        response = requests.get('http://www.seminarioabierto.com/metodos0{}.htm'.format(i))
    else: 
        response = requests.get('http://www.seminarioabierto.com/metodos{}.htm'.format(i))
        
    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    titulos.append(soup.find_all('table')[1].find('tr').find_all('td')[4].find('font').find('b').text)

titulos
```
